```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import glob

# Set up plotting style
plt.style.use("default")
sns.set_palette("husl")

# Define the header columns
header = [
    "name",
    "model",
    "id",
    "task",
    "var",
    "type",
    "num_steps",
    "success",
    "points",
    "golden",
]
```

```python
data_dir = Path("/home/natha/dev/tara_final_project/analysis/data")
csv_files = glob.glob(str(data_dir / "*.csv"))
print(csv_files)

all_data = []

for file_path in csv_files:
    file_name = Path(file_path).stem

    # Skip empty files
    if Path(file_path).stat().st_size == 0:
        print(f"Skipping empty file: {file_name}")
        continue

    try:
        df = pd.read_csv(file_path, header=None, names=header)
        all_data.append((file_name, df))
        print(f"Loaded {len(df)} records from {file_name}")
    except Exception as e:
        print(f"Error loading {file_name}: {e}")
```

```python
cols = []
def adjust_points(row):
    if row['success'] == True:
        return 100
    else:  # success == False
        if row['points'] < 0:
            return 0
        else:
            return row['points']


# Get sorted task names consistently
reference_df = all_data[0][1].copy().sort_values(by='task')
task_names = reference_df['task'].values

for thing in all_data:
    df = thing[1].copy()
    df['points'] = df.apply(adjust_points, axis=1)
    df = df.sort_values(by='task').reset_index(drop=True)
    cols.append(df['points'].rename(thing[0]))

graph = pd.concat(cols, axis=1)
graph.index = task_names
graph['Total'] = graph.sum(axis=1)
graph.sort_values(by='Total', ascending=False, inplace=True)
# Manually specify column order (replace with your actual column names)
desired_order = ['tips_sonnet4', 'default_sonnet4', 'human_sonnet4', 'default_r1', 
                 'high_flash', 'tips_flash', 'default_flash', 'default_deepseek', 
                 'higher_flash', 'nothing_flash', 'default_41mini']

# Reorder columns
graph_reordered = graph[desired_order]
graph_reordered['Total'] = graph_reordered.sum(axis=1)
graph_reordered
```

```python
# Reorder columns (exclude Total for heatmap)
available_columns = [col for col in desired_order if col in graph.columns]
graph_for_heatmap = graph[available_columns]

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(graph_for_heatmap, 
            annot=True,  # Show values in cells
            cmap='RdYlGn',  # Red-Yellow-Green colormap (red=low, green=high)
            center=50,  # Center the colormap at 50 points
            fmt='g',  # Format for annotations
            cbar_kws={'label': 'Points'},
            linewidths=0.5)

plt.title('Method Performance Heatmap by Task', fontsize=16, pad=20)
plt.xlabel('Methods', fontsize=12)
plt.ylabel('Tasks', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Alternative: Smaller heatmap with no annotations for cleaner look
plt.figure(figsize=(10, 8))
sns.heatmap(graph_for_heatmap, 
            annot=False,  # No text in cells
            cmap='viridis',  # Different colormap
            cbar_kws={'label': 'Points'})

plt.title('Method Performance Heatmap (Clean)', fontsize=14)
plt.xlabel('Methods')
plt.ylabel('Tasks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Show the data table as well
print("Data being visualized:")
print(graph_for_heatmap)
```

```python
import matplotlib.pyplot as plt

# Your existing code to create summary_df
temp = all_data[2][1]
df = temp
new_header = [
    "file_name",
    "model",
    "score",
    "wins",
    "win_percent"
]
new_data = []

def adjust_points(row):
    if row['success'] == True:
        return 100
    else:  # success == False
        if row['points'] < 0:
            return 0
        else:
            return row['points']

for entry in all_data:
    df = entry[1]
    name = entry[0]
    df['points'] = df.apply(adjust_points, axis=1)
    score = df['points'].mean()
    wins = df['success'].sum()
    win_percent = (wins / len(df)) * 100
    model = df['model'].iloc[1]
    new_entry = [name, model, score, wins, win_percent]
    new_data.append(new_entry)

summary_df = pd.DataFrame(new_data, columns=new_header)
summary_df.sort_values(by='score', ascending=False, inplace=True)

# Pretty horizontal bar chart with highest at top
plt.figure(figsize=(10, 8))

# Create horizontal bars with color gradient
colors = plt.cm.viridis(summary_df['score'] / summary_df['score'].max())
bars = plt.barh(summary_df['file_name'], summary_df['score'], color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)

# Add score labels on the bars
for i, (bar, score) in enumerate(zip(bars, summary_df['score'])):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, 
             f'{score:.1f}', ha='left', va='center', fontweight='bold', fontsize=10)

plt.xlabel('Average Score', fontsize=12, fontweight='bold')
plt.ylabel('Model', fontsize=12, fontweight='bold')
plt.title(' Model Performance Comparison', fontsize=16, fontweight='bold', pad=20)
plt.grid(axis='x', alpha=0.3, linestyle='--')
plt.xlim(0, 100)

# Reverse y-axis so highest score is at top
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
